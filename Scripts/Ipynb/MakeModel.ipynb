```python
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
## line 5 ~ 64 https://hwanny-yy.tistory.com/11 기반으로 작성
# 1. 분류에서 회귀로 수정
# 2. 이미지 사이즈 450 x 450


def build_vgg_block(input_layer,
                    num_cnn=3, 
                    channel=64,
                    block_num=1,
                   ):                   
    # 입력 레이어
    x = input_layer
	# num_cnn : 한블럭에서 사용할 conv필터 개수 네트워크에 따라 2개일때가 있고 3개일때가 있음.
    # CNN 레이어
    for cnn_num in range(num_cnn):
        x = tf.keras.layers.Conv2D(
            filters=channel,
            kernel_size=(3,3),
            activation='relu',
            kernel_initializer='he_normal',
            padding='same',
            name=f'block{block_num}_conv{cnn_num}'
        )(x)    

    # Max Pooling 레이어
    x = tf.keras.layers.MaxPooling2D(
        pool_size=(2, 2),
        strides=2,
        name=f'block{block_num}_pooling'
    )(x)

    return x

# 기본적으로 VGG-16으로 파라미터가 설정되어있다.
## 매개변수를 num_cnn_list=[2,2,4,4,4], channel_list=[64,128,256,512,512] 로 설정하면 VGG-19

def build_vgg(input_shape=(32,32,3),
              num_cnn_list=[2,2,3,3,3],
              channel_list=[64,128,256,512,512],
              num_classes=10):
    
    assert len(num_cnn_list) == len(channel_list) #모델을 만들기 전에 config list들이 같은 길이인지 확인합니다.
    
    
    
    
    input_layer = tf.keras.layers.Input(shape=input_shape)  # input layer를 만들어둡니다.
    output = input_layer
    
    # config list들의 길이만큼 반복해서 블록을 생성합니다.
    for i, (num_cnn, channel) in enumerate(zip(num_cnn_list, channel_list)):
        output = build_vgg_block(
            output,
            num_cnn=num_cnn, 
            channel=channel,
            block_num=i
        )
        
    output = tf.keras.layers.Flatten(name='flatten')(output)
    output = tf.keras.layers.Dense(4096, activation='relu', name='fc1')(output)
    output = tf.keras.layers.Dense(4096, activation='relu', name='fc2')(output)
    output = tf.keras.layers.Dense(136, activation='softmax', name='softmax')(output)
    
    model = tf.keras.Model(
        inputs=input_layer, 
        outputs=output
    )
    return model

testmodel = build_vgg(input_shape=(225,225,1))

testmodel.summary()
```

```python
testmodel.compile(
    optimizer='adam',
    loss='mse',
)

```

```python
history = testmodel.fit(
    train_images,
    validation_data=val_images,
    epochs=100,
    callbacks=[
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=5,
            restore_best_weights=True
        )
    ]
)
```
